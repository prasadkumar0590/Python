```python
class A:
    def __init__(self,id,name):
        self.id=id
        self.name=name
    def __str__(self):
        return f'Id:{self.id} Name:{self.name}'

    def MethodA(self):
        print('MethodA')

class B(A):

    def __init__(self,id,name,age):
        self.id=id
        self.name=name
        self.age=age
    def __str__(self):
        return super().__str__() + f' age:{self.age}'
    
    def MethodA(self):
        super().MethodA()
        print('MethodB')
        super().MethodA()

def run(obj):
    obj.MethodA()
a=A(1,'A')
b=B(2,'B','23');
print(a)
print(b)
print(run(a))
print(run(b))
```

```python
# Class, instance and static methods

class Employee:
    id=10
    Name='ABC'
    def __init__(self,var1,age):
        self.var1=var1
        self.age=age
        
    def print_instance_values(self):
        print('Var1: {} age : {}'.format(self.var1,self.age))
        
    @classmethod
    def print_class_values(cls):
        print('ID: {} Name : {}'.format(cls.id,cls.Name))
    @staticmethod
    def static_method():
        print('non Class & non instance method')
e=Employee('Test',28)
e.print_instance_values()
e.print_class_values()
print(Employee.id,Employee.Name)
print(e.id,e.Name)
Employee.static_method()

#e.print_instance_values()
```

```python
class Employee:
    count=0
    def __init__(self,name):
        self.__name=name
        Employee.count=Employee.count+1
    @property
    def name(self):
        return self.__name
    @name.setter
    def name(self,name):
        self.__name=name
    @classmethod
    def get_class_count(cls):
        return cls.count
    @staticmethod
    def gloabl_method():
        print('gloabl_method')

e1=Employee('Emp1')
e2=Employee('Emp2')
e1.name='ABC'
print(e1.name)
print(Employee.get_class_count())
print(e1.get_class_count())
print(e2.get_class_count())
print(Employee.gloabl_method())
```

```python
# generator 
# a function contains yeild keyword then it is a generator
# we can iterate the generator using for loop
# we can get elements by __next__()

def feb():
    a,b=0,1
    n=5
    while n!= 0:
        yield a
        a,b=b,a+b
        n-=1
fun=feb()
for x in range(5):
    print(fun.__next__())
    
```
