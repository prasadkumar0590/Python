```python
# find pairs whose sum is given number

def find_pairs(items,k):
    seen = set()
    for num in items:
        target = k-num
        if target not in seen:
            seen.add(num)
        else:
            yield (target,num)

for x in find_pairs([1,2,3,2],4):
    print(x)
```

```python
# Find common elements in two lists
def get_common_elements(l1,l2):
    i=j=0
    while(i < len(l1) and j < len(l2)):
        if l1[i] == l2[j]:
            yield l1[i]
            i +=1
            j +=1
        elif l1[i] < l2[j]:
            i +=1
        else:
            j +=1
l1=[1,2,3,4]
l2=[0,1,2]
for x in get_common_elements(l1,l2):
    print(x,end=" ")
```

```python
#Fibonaci Series
def print_fibonaci(n):
    a,b=0,1
    index = 0
    while(index < n):
        yield a
        a,b=a+b,a
        print(a,b)
        index +=1
for x in print_fibonaci(10):
    print(x)
```

```python
# decorators if we want change the behavior of existing method without writing any code in exsiting method we can do this with decorator
```

```python
def add_decorator(func):
    def inner(*args):
        if len(args) > 2:
            return False
        else:
            return func(*args)
    return inner
@add_decorator
def add(a,b):
    return a+b

value = add(1,2)
print(value)
```

```python
#Custom Iterator

class MyIterator:
    def __init__(self,max):
        self.max=max
    def __iter__(self):
        self.num = 0
        return self
    def __next__(self):
        if self.num < self.max:
            self.num +=1
            return self.num
        else:
            raise StopIteration
for x in MyIterator(10):
    print(x)
```

```python
import os, random, string
def generate_password():
    chars = string.ascii_letters + "!@#$%^&*"
    random.seed = os.urandom(1024)

    return "".join(random.choice(chars) for i in range(10)) + str(random.randint(0,9))

print(generate_password())
```

```python
# Multiple Inheritence follows Method Resolution Order, If we have same methods in parent & child classses then It will
# look which class is inherited first


class A:
    def a(self):
        print('aA')
class B:
    def a(self):
        super().a()
        print('aB')
class C(B,A):
    def c(self):
        print('c')
c = C()
c.a()
```
