```python
def logger(func):
    def logger_func(*args):
        print(func(*args))
    return logger_func

def add(x,y):
    return x+y
def sub(x,y):
    return x-y

add_logger=logger(add)
add_logger(3,4)
add_logger(10,20)

sub_logger=logger(sub)
sub_logger(20,10)
sub_logger(100,40)
```

# Decorators

```python
def decorator_demo(func):
    def wrapper(*args,**kwargs):
        print('this is wrapper. before calling original method')
        func(*args,**kwargs)
    return wrapper

@decorator_demo
def display_info():
    print('display_info()')
#disp_info=decorator_demo(display_info)
#disp_info()
display_info()
```

```python
import sys
from functools import wraps
def decorator_demo(func):
    @wraps(func)
    def wrapper(*args,**kwargs):
        print('this is wrapper. before calling original method')
        print(len(args))
        if len(args) > 1 and sum(args) > 20:
            raise Exception('sum can\'t be more than 20')   
        func(*args,**kwargs)
    return wrapper

@decorator_demo
def display_info():
    print('display_info()')
#disp_info=decorator_demo(display_info)
#disp_info()
@decorator_demo
def add(a,b):
    print('add: ',(a+b))

@decorator_demo
def test():
    print('test')
display_info()
add(2,3)
try:
   add(20,3) 
except Exception as e:
    print(e)
test()
```

# Generator

```python
def gen_demo(num):
    count=0;
    while num > count:
        yield count
        count+=1
for x in gen_demo(10):
    print(x)
```

# Recursion

```python
# Reverse a string

def reverse(string):
  if len(string) == 0:
    return string
  return reverse(string[1:])+string[0]
print(reverse('abcde'))
```

```python
def fact(num):
    if num == 1:
        return num
    return num*fact(num-1)
print(fact(5))
```

```python
def fib(a,b,num):
    if num==0:
        return ''
    print(a)
    a,b=b,a+b
    return fib(a,b,num-1)
print(fib(0,1,5))
```
