```
def outer(func):
  def inner(*args,**kwargs):
    print(args,kwargs)
    func(*args,**kwargs)
  return inner

@outer
def add(a,b):
  print(a+b)
add(1,2)
```

```
class IterDemo:
  def __init__(self):
    self.count=0
  def __iter__(self):
    return self
  def __next__(self):
    self.count=self.count+1
    if self.count > 10:
      raise StopIteration()
    return self.count
for x in IterDemo():
  print(x)
```

```
def Gen(a,b,num):
  while num >0:
    yield a
    a,b=b,a+b
    num-=1
for x in Gen(0,1,6):
  print(x)
  
```
